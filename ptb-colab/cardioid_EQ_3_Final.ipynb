```python
import numpy as np
import matplotlib.pyplot as plt

# ECG 구성 요소 함수 정의
def ecg_component(t, component_type):
    if component_type == "P":
        return 0.0413 * np.cos(t) + 0.0041 * (np.cos(t) ** 2) - 0.0016
    elif component_type == "QRS":
        return 0.6391 * np.cos(t) + 0.5636 * (np.cos(t) ** 2) - 0.1041
    elif component_type == "T":
        return 0.1391 * np.cos(t) + 0.0449 * (np.cos(t) ** 2) - 0.0155
    elif component_type == "base_line":
        return np.zeros_like(t)
    else:
        raise ValueError("Invalid component type. Choose from 'P', 'QRS', 'T', or 'base_line'.")

# 각 사이클의 ECG 신호 생성
def generate_ecg_cycle():
    f_b = np.linspace(0, f_base_line_duration, int(f_base_line_duration * sampling_rate))
    t_p = np.linspace(f_base_line_duration, f_base_line_duration + p_duration, int(p_duration * sampling_rate))
    t_pq = np.linspace(f_base_line_duration + p_duration, f_base_line_duration + p_duration + pq_duration, int(pq_duration * sampling_rate))
    t_qrs = np.linspace(f_base_line_duration + p_duration + pq_duration, f_base_line_duration + p_duration + pq_duration + qrs_duration, int(qrs_duration * sampling_rate))
    t_st = np.linspace(f_base_line_duration + p_duration + pq_duration + qrs_duration, f_base_line_duration + p_duration + pq_duration + qrs_duration + st_duration, int(st_duration * sampling_rate))
    t_t = np.linspace(f_base_line_duration + p_duration + pq_duration + qrs_duration + st_duration, f_base_line_duration + p_duration + pq_duration + qrs_duration + st_duration + t_duration, int(t_duration * sampling_rate))
    s_b = np.linspace(f_base_line_duration + p_duration + pq_duration + qrs_duration + st_duration + t_duration, total_time, int(s_base_line_duration * sampling_rate))

    x_fb = ecg_component(np.linspace(-np.pi, np.pi, len(f_b)), "base_line")
    x_p = ecg_component(np.linspace(-np.pi, np.pi, len(t_p)), "P")
    x_pq = ecg_component(np.linspace(-np.pi, np.pi, len(t_pq)), "base_line")
    x_qrs = ecg_component(np.linspace(-np.pi, np.pi, len(t_qrs)), "QRS")
    x_st = ecg_component(np.linspace(-np.pi, np.pi, len(t_st)), "base_line")
    x_t = ecg_component(np.linspace(-np.pi, np.pi, len(t_t)), "T")
    x_sb = ecg_component(np.linspace(-np.pi, np.pi, len(s_b)), "base_line")

    x_p_adjusted = x_p - x_p[-1]
    x_qrs_adjusted = x_qrs - x_qrs[-1]
    x_t_adjusted = x_t - x_t[-1]

    return np.concatenate([x_fb, x_p_adjusted, x_pq, x_qrs_adjusted, x_st, x_t_adjusted, x_sb])

# 설정값
sampling_rate = 1000  # 샘플링 주파수 (Hz)
total_time = 0.83  # 한 사이클의 총 시간 (초)
f_base_line_duration = 0.2
p_duration = 0.08
pq_duration = 0.075
qrs_duration = 0.08
st_duration = 0.15
t_duration = 0.2
s_base_line_duration = 0.045

# R-peak 시점 계산
# Parameters
HR = 1.83  # Heart rate in Hz (1 beat per second)
Cs, Cp = 0.8, 0.2  # Weights for sympathetic and parasympathetic signals
ws, wp = 2 * np.pi * 0.1, 2 * np.pi * 0.4  # Frequencies of neural oscillators
threshold = 1  # Threshold for R-peak generation
duration = 10  # Total duration for 3 cycles
dt = 0.001  # Time step

# Time vector
t = np.arange(0, duration, dt)

# Neural oscillators
s = np.sin(ws * t)  # Sympathetic signal
p = np.sin(wp * t)  # Parasympathetic signal
m = Cs * s + Cp * p  # Modulated signal

# IPFM model
integral = 0
r_peaks = []
for i in range(len(t)):
    integral += (HR + m[i]) * dt
    if integral >= threshold:
        r_peaks.append(t[i])
        integral = 0  # Reset integral

# RR intervals
rr_intervals = np.diff(r_peaks)  # R-peak 간의 차이를 계산
print("R-R interval:", rr_intervals)

r_peaks = np.cumsum(rr_intervals)  # RR 간격의 누적 합

# ECG 신호 생성
x_total = np.array([])
t_total = np.array([])

for i, r_peak_time in enumerate(r_peaks):
    x_cycle = generate_ecg_cycle()

    # 현재 RR 간격에 맞춰 사이클 시간 벡터 생성
    t_cycle = np.linspace(r_peak_time - rr_intervals[i], r_peak_time, len(x_cycle))

    x_total = np.concatenate([x_total, x_cycle])
    t_total = np.concatenate([t_total, t_cycle])

# 그래프 플롯
plt.figure(figsize=(12, 6))
plt.plot(t_total, x_total, label='ECG Waveform with IPFM Stress 3', color='blue')
plt.title('ECG Waveform with IPFM Stress 3', fontsize=14)
plt.xlabel('seconds', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
