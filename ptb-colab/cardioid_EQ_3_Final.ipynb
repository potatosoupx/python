
$$ ECG_{syn}=
\begin{Bmatrix}
  P(t)= ( x_p(t),y_p(t) ) & \text{if } t\in[0,\pi ] \\
  QRS(t)= (x_{qrs}(t), y_{qrs}(t)) & \text{if } t\in[\pi ,2\pi ] \\
  T(t)= (x_t(t),y_t(t)) & \text{if } t\in[2\pi,3\pi ]
\end{Bmatrix} $$

$$x_{p,qrs,t} = A\times \cos(t)+B\times \cos^2(t)+C$$
$$y_{p,qrs,t} = D\times \sin(t)+E\times \cos(t)\times \sin(t)+F$$
